```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

a = -5
b = 5
X = np.linspace(a, b, 100)
Y = 5-(X**4 + 2*X**3 - 3*X**2 +13)**0.5
plt.figure()
plt.plot(X, Y, 'k')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid()
plt.show()
```

```python
def Y(X):
  return 5-(X**4 + 2*X**3 - 3*X**2 +13)**0.5
```

```python
Aq = Y(-2.18)
print(Aq)
```

```python
def GoldenSectionSearch():
    a = -5
    b = 5
    tau = 2 - 1.618033988
    epsilon = 0.00001

    cont = 0
    registro = []

    while(True):
        # Calcular alpha1 y alpha2
        alpha1 = a*(1 - tau) + b*tau
        alpha2 = a*tau + b*(1 - tau)

        # Calcular f(alpha1) y f(alpha2)
        Y_alpha1 = Y(alpha1)
        Y_alpha2 = Y(alpha2)

        if(Y_alpha1 < Y_alpha2):
            a = alpha1
        else:
            b = alpha2

        cont = cont + 1
        registro.append([cont, alpha1, Y_alpha1])
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, alpha1, Y_alpha1))

        if(np.abs(Y_alpha1 - Y_alpha2) < epsilon):
            print("-------------------------------------------------------")
            print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, alpha1, Y_alpha1))
            break

    return registro
```

```python
reg1= GoldenSectionSearch()
```

##Evaluacin del mtodo##

Vamos a plotear el proceso de evolucin que el algoritmo sigui para encontrar el mximo .

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-2.187, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(4.220, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg1)
```
