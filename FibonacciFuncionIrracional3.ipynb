```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

a = -5
b = 5
X = np.linspace(a, b, 100)
Y = 5-(X**4 + 2*X**3 - 3*X**2 +13)**0.5
plt.figure()
plt.plot(X, Y, 'k')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid()
plt.show()
```

```python
def Y(X):
  return 5-(X**4 + 2*X**3 - 3*X**2 +13)**0.5
```

```python
def Fibonacci(*params):
    a   = params[0]
    b   = params[1]
    k   = params[2]
    it  = params[3]
    reg = params[4]
    L   = params[5]
    n   = params[6]
    fs  = params[7]
    ep  = params[8]

    i_arriba = n - k + 1
    i_abajo = n + 1
    Lk = L*fs[i_arriba]/fs[i_abajo]

    xa = a + Lk
    xb = b - Lk

    Y_xa = Y(xa)
    Y_xb = Y(xb)

    if(Y_xa < Y_xb):
        a = xa
    else:
        b = xb

    if(np.abs(Y_xa-Y_xb) > ep):
        k = k + 1
        it = it + 1

        reg.append([it, xa, Y_xa])
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(it, xa, Y_xa))

        return Fibonacci(a, b, k, it, reg, L, n, fs, ep)
    else:
        print("-------------------------------------------------------")
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(it, xa, Y_xa))

        return reg
```

```python
def RunFibonacci():
    n = 30

    fs = [1, 1]
    for i in range(n):
        aux = fs[-1] + fs[-2]
        fs.append(aux)

    a, b, k = -5, 5, 2
    L = b - a

    return Fibonacci(a, b, k, 0, [], b-a, n, fs, 0.0001)
```

```python
reg1 = RunFibonacci()
```

##Evaluacion del metodo##

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-2.187, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(4.220, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg1)
```
