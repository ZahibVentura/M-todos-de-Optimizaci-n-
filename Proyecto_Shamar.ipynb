```python
from os import XATTR_SIZE_MAX
X = sum(años)
Y = sum(demanda)
XY = X*Y
A = [0,0]
for i in range(len(años)):
  A[0] += años[i]*demanda[i]
  A[1] += años[i]**2

a = (len(años)*A[0] - XY)/(len(años)*A[1] - X**2)
b = (Y - a*X)/len(años)
x_m = np.mean(años)
y_m = np.mean(demanda)

d = 0
d1 = 0
d2 = 0
for i in range(len(años)):
  d += (años[i]-x_m)*(demanda[i]-y_m)
  d1 += (años[i]-x_m)**2
  d2 += (demanda[i]-y_m)**2
r = d/np.sqrt(d1*d2)


print(r)
print(a,b)
```
