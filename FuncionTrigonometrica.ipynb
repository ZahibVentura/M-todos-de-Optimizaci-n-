```python
def d2Y(X):
  return -218.7*(np.sin(9*X))**3 + 437.4*np.sin(9*X)*(np.cos(9*X))**2 + 60*np.sin(2*X - 30)*np.cos(15*X + 50) - 18*(np.sin(3*X + 45))**2 + 229*np.sin(15*X + 50)*np.cos(2*X - 30) + 18*(np.cos(3*X + 45))**2
```

```python
def Y(X):
  return 0.9*(np.sin(9*X))**3 + (np.sin(3*X + 45))**2 - np.sin(15*X + 50)*np.cos(2*X - 30)
```

```python
def dY(X):
  return 24.3*(np.sin(9*X))**2 *np.cos(9*X) + 2*np.sin(2*X - 30)*np.sin(15*X + 50) + 6*np.sin(3*X + 45)*np.cos(3*X + 45) - 15*np.cos(2*X - 30)*np.cos(15*X + 50)
```

```python
 # Mtodo de biseccion en python
import numpy as np
def biseccion():
    a = -5
    b = 0
    epsilon = 0.001

    cont = 0
    registro = []

    while(True):
        alpha = (a+b)/2

        # Primera derivada en a
        dY_a = dY(a)

        #Primera derivada en alpha
        dY_alpha = dY(alpha)

        if(dY_a*dY_alpha < 0):
            b = alpha
        else:
            a = alpha
        Ya = Y(a)

        # Condicion de finalizacion
        if(np.abs(a-b) < epsilon):
            print('-------------------------------------------------------------')
            print("It: {:02} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, a, Ya))
            break

        cont = cont + 1
        registro.append([cont, a, Ya])
        print("It: {:02} - Eje X: {:.10f} - Eje X: {:.10f}".format(cont, a, Ya))

    return registro


```

```python
reg1 = biseccion()
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-1.944, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(2.54, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg1)
```

##Secante##


```python
def Secante():
    a= -5
    b= 0
    epsilon = 0.001

    cont = 0
    registro = []

    while True:
        # calcular f'(a)
        dY_a = dY(a)

        # calcular f'(b)
        dY_b= dY(b)

        # calcular alpha
        alpha = b - dY_b / ((dY_b - dY_a)/(b - a))

        dY_alpha = dY(alpha)

        if dY_alpha > 0:
            b = alpha
        else:
            a = alpha

        Y1_alpha = Y(alpha)

        cont = cont + 1

        registro.append([cont, alpha, Y1_alpha])
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, alpha, Y1_alpha))

        if np.abs(dY_alpha) <= epsilon:
            print("------------------------------------------------------------------------")
            print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, alpha, Y1_alpha))
            break

    return registro
```

```python
reg2 = Secante()
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-1.944, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(2.54, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg2)
```

##Metodo de NewtonRaphson##

```python
def NewtonRaphson():
    a = -5
    b = 0
    x = -1
    epsilon = 0.001

    cont = 0
    registro = []

    while True:
        # Calculo de f'(x) y f''(x)
        dY_x = dY(x)
        d2Y_x = d2Y(x)

        xprev = x
        x = xprev - dY_x/d2Y_x

        Y_x = Y(x)

        cont = cont + 1
        registro.append([cont, x, Y_x])

        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, x, Y_x))

        if(np.abs(x - xprev) <= epsilon):
            print("----------------------------------------------------------------")
            print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, x, Y_x))
            break

    return registro
```

```python
reg3 = NewtonRaphson()
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-1.944, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(2.54, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg3)
```

##Seccion Dorada##

```python
def GoldenSectionSearch():
    a = -5
    b = 0
    tau = 2 - 1.618033988
    epsilon = 0.00001

    cont = 0
    registro = []

    while(True):
        # Calcular alpha1 y alpha2
        alpha1 = a*(1 - tau) + b*tau
        alpha2 = a*tau + b*(1 - tau)

        # Calcular f(alpha1) y f(alpha2)
        Y_alpha1 = Y(alpha1)
        Y_alpha2 = Y(alpha2)

        if(Y_alpha1 < Y_alpha2):
            a = alpha1
        else:
            b = alpha2

        cont = cont + 1
        registro.append([cont, alpha1, Y_alpha1])
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, alpha1, Y_alpha1))

        if(np.abs(Y_alpha1 - Y_alpha2) < epsilon):
            print("-------------------------------------------------------")
            print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, alpha1, Y_alpha1))
            break

    return registro
```

```python
reg4 = GoldenSectionSearch()
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-1.944, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(2.54, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg4)
```

##Fibonacci##

```python
def Fibonacci(*params):
    a   = params[0]
    b   = params[1]
    k   = params[2]
    it  = params[3]
    reg = params[4]
    L   = params[5]
    n   = params[6]
    fs  = params[7]
    ep  = params[8]

    i_arriba = n - k + 1
    i_abajo = n + 1
    Lk = L*fs[i_arriba]/fs[i_abajo]

    xa = a + Lk
    xb = b - Lk

    Y_xa = Y(xa)
    Y_xb = Y(xb)

    if(Y_xa < Y_xb):
        a = xa
    else:
        b = xb

    if(np.abs(Y_xa-Y_xb) > ep):
        k = k + 1
        it = it + 1

        reg.append([it, xa, Y_xa])
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(it, xa, Y_xa))

        return Fibonacci(a, b, k, it, reg, L, n, fs, ep)
    else:
        print("-------------------------------------------------------")
        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(it, xa, Y_xa))

        return reg
```

```python
def RunFibonacci():
    n = 30

    fs = [1, 1]
    for i in range(n):
        aux = fs[-1] + fs[-2]
        fs.append(aux)

    a, b, k = -5, 0, 2
    L = b - a

    return Fibonacci(a, b, k, 0, [], b-a, n, fs, 0.0001)

```

```python
reg5 = RunFibonacci()
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(-1.944, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Eje X')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(2.54, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Eje Y')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg5)
```
