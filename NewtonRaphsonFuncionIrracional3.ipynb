```python
# FUNCION ORIGINAL
def Y(X):
  return 5 - (X**4 + 2*X**3 - 3*X**2 + 13)**0.5
```

```python
# DERIVADA DE LA FUNCION
def dY(X):
  return -(2.0*X**3 + 3.0*X**2 - 3.0*X)/(X**4 + 2*X**3 - 3*X**2 + 13)**0.5
```

```python
# 2da DERIVADA DE LA FUNCION
def d2Y(X):
  return -(6.0*X**2 + 6.0*X - 3.0)/(X**4 + 2*X**3 - 3*X**2 + 13)**0.5 - (-2.0*X**3 - 3.0*X**2 + 3.0*X)*(2.0*X**3 + 3.0*X**2 - 3.0*X)/(X**4 + 2*X**3 - 3*X**2 + 13)**1.5
```

```python
def NewtonRaphson():
    a = -5
    b = 5
    x = -1
    epsilon = 0.001

    cont = 0
    registro = []

    while True:
        # Calculo de f'(x) y f''(x)
        dY_x = dY(x)
        d2Y_x = d2Y(x)

        xprev = x
        x = xprev - dY_x/d2Y_x

        Y_x = Y(x)

        cont = cont + 1
        registro.append([cont, x, Y_x])

        print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, x, Y_x))

        if(np.abs(x - xprev) <= epsilon):
            print("----------------------------------------------------------------")
            print("It: {:02d} - Eje X: {:.10f} - Eje Y: {:.10f}".format(cont, x, Y_x))
            break

    return registro
```
